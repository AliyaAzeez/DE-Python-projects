### Database Set Up function

```python
def create_database():
    
    # creating a connection to default database
    conn = psycopg2.connect("host=127.0.0.1 dbname =postgres user= postgres password=")
    cur = conn.cursor()
    conn.set_session(autocommit=True)
    
    # create a new database
    cur.execute("DROP DATABASE IF EXISTS wealth")
    cur.execute("CREATE DATABASE wealth")
    
    #close connection to the base database
    cur.close()
    
    #connect to the new database
    conn = psycopg2.connect("host=127.0.0.1 dbname=wealth user=postgres password=")
    cur = conn.cursor()
    
    return cur, conn
```

```python
cur,conn = create_database()
```

```python
cur.execute("CREATE TABLE IF NOT EXISTS wealth1995 (Economy varchar,Code varchar,Region varchar,IncomeGr varchar, Population varchar, TotalWealth varchar); ")
conn.commit()
```

```python
cur.execute("CREATE TABLE IF NOT EXISTS wealth2000 (Economy varchar,Code varchar,Region varchar,IncomeGr varchar, Population varchar, TotalWealth varchar); ")
conn.commit()
```

```python
cur.execute("CREATE TABLE IF NOT EXISTS wealth2005 (Economy varchar,Code varchar,Region varchar,IncomeGr varchar, Population varchar, TotalWealth varchar); ")
conn.commit()
```

```python
wealth1998_insert = ("""INSERT INTO wealth1995 (Economy,Code,Region,IncomeGr,Population,TotalWealth) VALUES (%s, %s, %s, %s, %s, %s)""")
```

```python
for i,row in wealth1995.iterrows():
    cur.execute(wealth1998_insert,list(row))
```

```python
conn.commit()
```
